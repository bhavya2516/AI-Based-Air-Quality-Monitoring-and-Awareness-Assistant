```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_phishing.csv")
```

```python
print(df.head())
print(df.columns)
print(df.isnull().sum())
```

```python
import re
from urllib.parse import urlparse

def having_ip(url):
    return 1 if re.search(r'\d+\.\d+\.\d+\.\d+', url) else 0

def url_length(url):
    return len(url)

def count_dots(url):
    return url.count('.')

def has_at(url):
    return 1 if '@' in url else 0

def has_hyphen(url):
    return 1 if '-' in url else 0

def https_present(url):
    return 1 if urlparse(url).scheme == 'https' else 0

# Apply feature extraction
df['ip_present'] = df['url'].apply(having_ip)
df['url_length'] = df['url'].apply(url_length)
df['dot_count'] = df['url'].apply(count_dots)
df['at_symbol'] = df['url'].apply(has_at)
df['hyphen'] = df['url'].apply(has_hyphen)
df['https'] = df['url'].apply(https_present)

print(df.head())
```

```python
df = df.drop([
    'ip_present',
    'url_length',
    'dot_count',
    'at_symbol',
    'hyphen',
    'https'
], axis=1)
```

```python
X = df.drop(['url', 'status'], axis=1)
y = df['status']

print("Features shape:", X.shape)
print("Labels shape:", y.shape)
```

```python
processed_df = pd.concat([X, y], axis=1)
processed_df.to_csv("processed_phishing_milestone1.csv", index=False)

print("Milestone-1 preprocessing completed successfully")
```
